# 4. Logistic Regression

Since we had problems when trying to get a logistic regression model because it reached the maximum number of iterations, we will create the model with a subset of the columns (that is, with the most important features).

We have different series containing the important features for each model created with the RF and SVC models.

```python
import pandas as pd

# Data to be used for the model:
xtrain = pd.read_csv("xtrain.csv", index_col=0)
ytrain = pd.read_csv("ytrain.csv", index_col=0)

# Data to be used for the predictions:
xtest = pd.read_csv("xtest.csv", index_col=0)
ytest = pd.read_csv("ytest.csv", index_col=0)
```

## a) LR with the top 15 important features
### a1) from RF

```python
import pickle
from joblib import dump, load
features_RFC_15 = load("important_features_RFC_15.joblib")
features_RFC_15 #series
```

```python
def MultipleColumns_model(xtrain, ytrain, c, estimator, series_imp_features):
    """Torna una llista amb els models creats i adaptats per a cada columna.
        Entrades: ytrain i número de columnes que agafem.
    """
    models = [] #llista dels models per a cada columna
    for i in range(c): #from 0 to number of columns
        models.append(estimator()) 
        models[i].fit(xtrain.iloc[:, series_imp_features[i].tolist()], ytrain.iloc[:,i]) #xtrain now will be a subset (just the imp features) 
        print("model", i, "done")
    return models
```

```python
from sklearn.linear_model import LogisticRegression as LR

# Creating the model
LRmodel = MultipleColumns_model(xtrain, ytrain, ytrain.shape[1], LR, features_RFC_15) 
```

```python
# Saving the model
dump(LRmodel, 'output/4a1_model.joblib')
```

```python
#new dataframe for saving the predictions
proba_pred = pd.DataFrame(columns=ytest.columns)
name_col = ytest.columns.tolist()

for i in range(ytest.shape[1]):
    proba_pred[name_col[i]] = LRmodel[i].predict_proba(xtest.iloc[:, features_RFC_15[i].tolist()])[:, 1]

proba_pred["sig_id"]= ytest.index.tolist() #index, we want to know the sample ids
proba_pred = proba_pred.set_index('sig_id')

#Saving the probabilities:
proba_pred.to_csv(r'output/4a1_probas.csv')
```

## c) LR for each dosage-time combination

In this case, we'll be using the features: `list_imp_features_dosetime.joblib`, top 15 (which resulted in the best logistic loss score). 

```python
features_DoseTime = []
targets_DoseTime = []
dataset_DoseTime_names = []

for dose, time in zip([0,0,0,1,1,1], [24,48,72,24,48,72]):
    print('D', dose, 'T', time)
    features_DoseTime.append(pd.read_csv('dataDoseTime/xtrain_D'+str(dose)+'T'+str(time)+'.csv', index_col=0))
    targets_DoseTime.append(pd.read_csv('dataDoseTime/ytrain_D'+str(dose)+'T'+str(time)+'.csv', index_col=0))
    dataset_DoseTime_names.append("Dose"+str(dose)+"-Time"+str(time))
```

```python
#Which columns will give us problems? We can't have a column with just one class
print('\tCOLUMNS WITHOUT ANY 1\n')
i = 0
for targets in targets_DoseTime:
    for column in targets:
        if (1 in targets[column].tolist()) == False:
            print(column, '\n', dataset_DoseTime_names[i], '\n')
    i+=1
```

```python
#Creating a series data structure to save the lists of the columns with which a new model can be created for
#each combination:
cols_datasets_series = pd.Series(index=dataset_DoseTime_names, dtype=object)

#Creating the lists and filling the series:
i = 0
for targets in targets_DoseTime:
    cols_list = []
    for column in targets:
        if (1 in targets[column].tolist()) == True:
            cols_list.append(column)
    cols_datasets_series[dataset_DoseTime_names[i]] = cols_list
    i+=1
```

```python
#We want to make sure that all the columns of this dataset (0-48) will not give us problems when creating the models:
list_condition = []
for column in targets_DoseTime[1]:
    list_condition.append(1 in targets_DoseTime[1][column].tolist())
    
False in list_condition #if true: there are columns with no 1s
```

```python
#Modification!! 
def MultipleColumns_model_series(xtrain, ytrain, s, estimator, imp_features):
    """Return: Series modified with the name of the cols as keys and the models as values
        Input: xtrain, ytrain, series object, estimator
    """
    i = 0
    for c in s.index.tolist(): #names of the columns that have 1s 
        model = estimator() 
        s[c] = model.fit(xtrain.iloc[:, imp_features[c].tolist()], ytrain[c])
        print(i, end=' ', flush=True)
        i+=1
    return s
```

```python
# Creating the series, where the models will be saved:
model024 = pd.Series(index = cols_datasets_series['Dose0-Time24'], dtype=object)
model048 = pd.Series(index = cols_datasets_series['Dose0-Time48'], dtype=object)
model072 = pd.Series(index = cols_datasets_series['Dose0-Time72'], dtype=object)
model124 = pd.Series(index = cols_datasets_series['Dose1-Time24'], dtype=object)
model148 = pd.Series(index = cols_datasets_series['Dose1-Time48'], dtype=object)
model172 = pd.Series(index = cols_datasets_series['Dose1-Time72'], dtype=object)
```

```python
from joblib import load, dump
list_imp_features = load("list_imp_features_dosetime.joblib") #list where each element is a series containing
#the important features of every MoA model (6 elements --> 6 dose-time combinations)
```

```python
import time
#Remember that we always have the same order: dose 0-24,48,72 and dose 1-24, 48, 72
start = time.process_time()
i=0
for s, f, t, imp_f in zip([model024, model048, model072, model124, model148, model172], features_DoseTime, targets_DoseTime, list_imp_features):
    print("\tCreating the models for", dataset_DoseTime_names[i])
    s = MultipleColumns_model_series(f, t, s, LR, imp_f)
    print("Completed")
    i+=1
end = time.process_time()

total = end-start
print("Total execution time:", total)
```

```python
list_models_DoseTime = [model024, model048, model072, model124, model148, model172]
import pickle
from joblib import dump, load
model_saved = pickle.dumps(list_models_DoseTime)
dump(list_models_DoseTime, 'output/4c_model.joblib') 
```

**Predictions**

```python
#test datasets obtained from the different combinations and removing control samples:
ftest_tr = xtest[xtest['cp_type'] == 0] #only the treated samples (cp_type = 0)

ftest_DoseTime = []
ftest_DoseTime_names = [] #just in case we want to know the combination of dose and time that each dataframe has been created with

for dose in sorted(ftest_tr.cp_dose.unique().tolist()):
    for time in sorted(ftest_tr.cp_time.unique().tolist()):
        ftest_DoseTime.append(ftest_tr[(ftest_tr['cp_dose'] == dose) & (ftest_tr['cp_time'] == time)])
        ftest_DoseTime_names.append("Dose"+str(dose)+"-Time"+str(time))
```

```python
def prediction_given_models(targets, ftest_DoseTime_dataset, s, imp_features):
    
    class_pred = pd.DataFrame(columns=targets.columns)
    proba_pred = pd.DataFrame(columns=targets.columns)

    name_col = targets.columns.tolist()
    for col in name_col:
        if col in s.index.tolist(): #if there is a model created for that column
            #class_pred[col] = s[col].predict(ftest_DoseTime_dataset)
            proba_pred[col] = s[col].predict_proba(ftest_DoseTime_dataset.iloc[:,imp_features[col]])[:, 1]
        else: #if the column does not have a model associated with
            #class_pred[col] = [0 for i in range(ftest_DoseTime_dataset.shape[0])] #list with nrows zeros
            proba_pred[col] = [1/(10*ftest_DoseTime_dataset.shape[0]) for i in range(ftest_DoseTime_dataset.shape[0])]
            
    proba_pred["sig_id"]= ftest_DoseTime_dataset.index.tolist()
    proba_pred = proba_pred.set_index('sig_id')
    
    return proba_pred #dataframe
```

```python
probabilities = pd.DataFrame(columns=targets.columns)

for s, ftest_DoseTime_dataset, imp_f in zip(list_models_DoseTime, ftest_DoseTime, list_imp_features):
    prob_model = prediction_given_models(targets, ftest_DoseTime_dataset, s, imp_f)
    probabilities = pd.concat([probabilities, prob_model])
    
probabilities #all the treated samples predictions 
```

```python
##Now, we have all the treated samples with their probabilities
##However, the probabilities dataframe must include all the samples (also the control type)

ftest_ctrl = xtest[xtest['cp_type'] == 1] #only the control samples (cp_type = 1)

proba_pred_ctrl = pd.DataFrame(columns=targets.columns)
name_col = targets.columns.tolist()
for col in name_col:
    proba_pred_ctrl[col] = [0.0 for i in range(ftest_ctrl.shape[0])]
    
proba_pred_ctrl["sig_id"]= ftest_ctrl.index.tolist()
proba_pred_ctrl = proba_pred_ctrl.set_index('sig_id')

proba_pred_ctrl #all 0.0s
```

```python
#Finally, we'll have to concatenate this last dataframe with the other probabilities
all_samples_prediction = pd.concat([probabilities, proba_pred_ctrl])
all_samples_prediction = all_samples_prediction.reindex(ytest.index) #same order as ytest!
all_samples_prediction
```

```python
#Saving the probabilities:
all_samples_prediction.to_csv(r'output/4c_probas.csv')
```
